### SGDclassifier is efficent work on the large amount of independent variable  data .

```python
from sklearn.linear_model import SGDClassifier
sdg_classifier=SGDClassifier(random_state=42)
```

```python
sdg_classifier.fit(X_train,y_train_5)
```

so, We are Putting Data into the Prediction . it just training Data

```python
sdg_classifier.predict([some_digit])
```

* lets self Model Accuressy check
* After this put the Random values **(Test set into the Model)** & check What out Put Come.


```python
SDG_PREDICT_VALUES=sdg_classifier.predict(X_test)
print(f'no. of correct values predict by over model {sum(SDG_PREDICT_VALUES==y_test_5)}   |   out of  {len(y_test_5)}')
print(sum(SDG_PREDICT_VALUES==y_test_5)/len(y_test_5))
```

* Here we are checking the Model Accuressy  with the help of **Sklearn**
## CROSS VALIDATION 

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sdg_classifier,X_train,y_train_5,cv=3,scoring='accuracy')
```

##### **More control on Cross-Validation**

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone
```

```python
skfolds=StratifiedKFold(n_splits=3,random_state=42)     # Suffle Data is very Important Role play on Model Accuressy
```

#### Split Date into Three parts 
#### we are DO here model clone 

```python
for train_index, test_index in skfolds.split(X_train,y_train_5):   # three time iteration here done.
    clone_clf=clone(sdg_classifier)         # clone the model 
    X_train_fold=X_train[train_index]
    Y_train_fold=y_train_5[train_index]

    x_test_fold=X_train[test_index]
    y_test_fold=y_train_5[test_index]

    clone_clf.fit(X_train_fold,Y_train_fold)
    y_predict_fold =clone_clf.predict(x_test_fold)
    n_correct=sum(y_predict_fold==y_test_fold)    # Model predictive values & commpare with real test values
    print(f'no. of the correct values {n_correct}    |   out of {len(y_test_fold)}')
    print(n_correct/len(y_predict_fold))   #   you can multiply by 100 make it percentage
```

# What is **Model accuressy**
>>>> Model get Predictive values
>>>>> compare **Predictive values** with **real test values**  ` How many Correct values` & divided the len of data set
